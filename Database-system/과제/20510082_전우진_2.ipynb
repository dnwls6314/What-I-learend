```python
i=0
while i != 1000001:
    doc_1 = open("./temp.txt")
    data = doc_1.readline()
    temp_num = int(data)
    doc_1.close()
    
    doc_2 = open("./temp.txt", "r+")
    add_num = temp_num + 1
    add_num = str(add_num)
    doc_2.write(add_num)
    doc_2.close()
    i+=1
    
    if i%10000 == 0:
        print("{}/1000000".format(i))       
```
