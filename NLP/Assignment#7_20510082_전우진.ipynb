http://media.daum.net/digital 에서 뉴스 리스트와 URL을 불러옴

```python
#package load
import pandas as pd
import requests
from selenium import webdriver
import time
import random
from tqdm import tqdm
import requests
#pip install request
#pip install beautifulsoup4
from bs4 import BeautifulSoup
```

```python
#driver load
driver = webdriver.Chrome('./chromedriver.exe')

#url.get
url="http://media.daum.net/digital"
driver.get(url)
```

```python
html_source = driver.page_source
```

```python
html_source
```

```python
# 많이 본 뉴스 1~10위의 url
urls1 = []
for i in range(1,11):
    news_url = driver.find_element_by_xpath('//*[@id="mAside"]/div[1]/ul/li[1]/div/ol[1]/li['+ str(i) +']/strong/a')
    
    url = news_url.get_attribute('href')
    
    urls1.append(url)
```

```python
urls1
```

```python
# 많이 본 뉴스 11~20위의 url
urls2 = []
for i in range(1,11):
    news_url = driver.find_element_by_xpath('//*[@id="mAside"]/div[1]/ul/li[1]/div/ol[2]/li['+ str(i) +']/strong/a')
    
    url = news_url.get_attribute('href')
    
    urls2.append(url)
```

```python
urls2
```

```python
# 많이 본 뉴스 21~30위의 url
urls3 = []
for i in range(1,11):
    news_url = driver.find_element_by_xpath('//*[@id="mAside"]/div[1]/ul/li[1]/div/ol[2]/li['+ str(i) +']/strong/a')
    
    url = news_url.get_attribute('href')
    
    urls3.append(url)
```

```python
urls1.extend(urls2)
urls1.extend(urls3)
```

```python
len(urls1)
```

```python
# 기사 제목 추출
from urllib.request import urlopen
from bs4 import BeautifulSoup

titles=[]

for a in urls1:
    html = urlopen(str(a))  
    soup = BeautifulSoup(html, "html.parser") 
    title = soup.find('h3',{'class':'tit_view'}).text.strip()
    titles.append(title)
```
